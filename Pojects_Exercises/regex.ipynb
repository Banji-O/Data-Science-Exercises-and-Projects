### Extract financial periods from a company's financial reporting

```python
text = """
The gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
In previous quarter i.e. FY2020 Q4 it was $3 billion.
"""

pattern = 'FY\d{4} Q[1-4]'
matches = re.findall(pattern, text)
matches 
```

```python
text = """
The gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
In previous quarter i.e. fy2020 Q4 it was $3 billion.
"""

pattern = 'FY\d{4} Q[1-4]'
matches = re.findall(pattern, text, flags=re.IGNORECASE)
matches 
```

```python
text = """
The gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
In previous quarter i.e. fy2020 Q4 it was $3 billion.
"""

pattern = 'FY(\d{4} Q[1-4])[^\$]+\$([0-9\.]+)'
matches = re.findall(pattern, text, flags=re.IGNORECASE)
matches 
```

```python
text = """
The gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
In previous quarter i.e. fy2020 Q4 it was $3 billion.
"""

pattern = 'FY(\d{4} Q[1-4])[^\$]+\$([0-9\.]+)'
matches = re.search(pattern, text, flags=re.IGNORECASE)
matches 
```

Extract all Twitter handles from the following text. Twitter handle is the text that appears after https://twitter.com/ and is a single word. Also, it contains only alphanumeric characters i.e. A-Z a-z, 0 to 9, and underscore.

```python
text = """
Follow our leader Elon Musk on twitter here: https://twitter.com/elonmusk, more information on Teslas's products can be found at https://www.tesla.com/. Also here are leding influencers for tesla related news,
https://twitter.com/teslarati
https://twitter.com/dummy_tesla
https://twitter.com/_2_tesla
"""

pattern = r"https://twitter\.com/([A-Za-z0-9_]+)"

matches = re.findall(pattern, text)
matches
```
