<center><h1><b>Activation Function</b></h1></center>

This is method of using sigmoid or tanh to activate the output value. sigmoid converts class outputs to 0 - 1 values in binary classification, while tanh outputs are -1 to +1. This help to convert the output to percentage(0.003, 0.45, 0.89, etc) and the model with pick the maximum value. If not binary, tanh is always better in most cases. RelU was introduced because of the tendency of slow learning that might occur with sigmoid or tanh. 
The rule of thumb is that for hidden layers, if you are not sure of the activation function to use, just use ReLU as default choice. There is also a function similar to ReLU known as Leaky ReLU.

```python
import math

# This sigmoid function will convert value between 0 and 1
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(0.5)
```

```python
# This tanh function will convert value between -1 and 1
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

```python
#This ReLUfunction will take max between 0 and x. If the value is negative it returns 0 but will return the number if the value is positive
def relu(x):
    return max(0,x)
```

```python
relu(-7)
```

```python
relu(7)
```

```python
relu(8)
```

```python
def leaky_relu(x):
    return max(0.1 * x, x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
