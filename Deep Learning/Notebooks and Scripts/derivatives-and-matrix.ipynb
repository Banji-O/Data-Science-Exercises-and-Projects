### Dot Product

```python
price_per_unit = np.array([1000, 400, 1200])
units = np.array([[30, 40, 50], [5, 10, 15], [2, 5, 7]])
```

```python
# The equation below will make numpy to perform Broadcasting function 
total_revenue = price_per_unit * units
total_revenue
```

```python
# The dot product calculation
total_revenue = np.dot(price_per_unit, units)
total_revenue
```
