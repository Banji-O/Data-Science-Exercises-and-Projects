## Mean Absolute Error

```python
import numpy as np

y_predicted = np.array([1, 1, 0, 0,1])
y_true = np.array([0.30, 0.7, 1, 0, 0.5])
```

```python
# Mean Absolute Error Function
def mean_absolute_error(y_true, y_predicted):
    total_error = 0
    for y_tru, y_pred in zip(y_true, y_predicted):  # Zip is used to run for loop in 2 arrays at the same time
        total_error += abs(y_tru - y_pred)
    print(f'Total error: {total_error}')
    mean_absolute_error = total_error / len(y_true)
    print(f'Mean Absolute Error: {mean_absolute_error}')
    return mean_absolute_error
```

```python
mean_absolute_error(y_true, y_predicted)
```

### Using Numpy for the same operation

```python
mean_abs_error = np.mean(np.abs(y_predicted - y_true))
mean_abs_error
```

```python
total_error = np.sum(np.abs(y_predicted - y_true))
total_error
```

### Mean Square Error

```python
y_predicted = np.array([1, 1, 0, 0,1])
y_true = np.array([0.30, 0.7, 1, 0, 0.5])
```

```python
def mean_square_error(y_true, y_predicted):
    total_error = 0
    for y_tru, y_pred in zip(y_true, y_predicted):  # Zip is used to run for loop in 2 arrays at the same time
        total_error += (y_tru - y_pred) ** 2  # Square the difference to calculate squared error
    print(f'Total error: {total_error}')
    mean_square_error = total_error / len(y_true)  # taking average
    print(f'Mean Square Error: {mean_square_error}')
    return mean_square_error
```

```python
mean_square_error(y_true, y_predicted)
```

```python
meann_square_error = np.mean(np.square(y_predicted - y_true))
meann_square_error
```

## Log Loss or Binary Cross Entropy

```python
np.log([0.01])
```

```python
epsilon = 1e-15
epsilon
```

```python
y_predicted = np.array([1, 1, 0, 0,1])
y_true = np.array([0.30, 0.7, 1, 0, 0.5])
```

```python
y_predicted_new = [max(i, epsilon) for i in y_predicted]
y_predicted_new
```

```python
# Clip predicted values between epsilon and 1 - epsilon
y_predicted_new = np.clip(y_predicted, epsilon, 1 - epsilon)
y_predicted_new
```

```python
# Log Loss calculation
log_loss = -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))

log_loss
```

```python
# Putting the log loss in a function

def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    # Clip the predicted values between epsilon and 1 - epsilon
    y_predicted_new = np.clip(y_predicted, epsilon, 1 - epsilon)
    # Log loss calculation
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))


if __name__ == "__main__":
    y_true = np.array([0.30, 0.7, 1, 0, 0.5])
    y_predicted = np.array([1, 1, 0, 0, 1])

    print("Log Loss:", log_loss(y_true, y_predicted))
```

```python
log_loss(y_true, y_predicted)
```
