```python
import re
```

```python
chat1 = 'codebasics: you ask lot of questions 1235678912, abcc@xyz.com'
chat2 = 'codebasics: here it is: (123)-567-8912, abX_82@xyz.com'
chat3 = 'codebasics: yes, phone: 1235678912 email: abc@xyz.com' 
```

```python
pattern = r'\d{10}'

for text in [chat1, chat2, chat3]:
    matches = re.findall(pattern, text)
    print(matches)
```

```python
pattern = '\d{10}|\(\d{3}\)-\d{3}-\d{4}'

for text in [chat1, chat2, chat3]:
    matches = re.findall(pattern, text)
    print(matches)
```

```python
pattern = '[a-z0-9A-Z_]*@[a-z0-9A-Z]*\.\w*'

for text in [chat1, chat2, chat3]:
    matches = re.findall(pattern, text)
    print(matches[0])
```

```python
chat1 = 'codebasics: Hello, I am having an issue with my order # 412889912'
chat2 = 'codebasics: I have a problem with my order number 412889912'
chat3 = 'codebasics: My order 412889912 is having an issue, I was charged 300$ when online it says 280$'
```

```python
pattern = 'order[^\d]*(\d*)'

for text in [chat1, chat2, chat3]:
    matches = re.findall(pattern, text)
    print(matches[0])
```

```python
# Extract the name
pattern = "Born(.*)\n"

matches = re.findall(pattern, text)
matches[0].strip()
```

```python
 def get_pattern_match(pattern, text):
     matches = re.findall(pattern, text)
     if matches:
         return matches[0]
```

```python
get_pattern_match("\(age.*\n(.*)", text)
```

```python
def get_personal_information(text):
    age = get_pattern_match('age (\d+)', text)
    full_name = get_pattern_match('Born(.*)\n', text)
    birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
    birth_place = get_pattern_match('\(age.*\n(.*)', text)

    return {"Full Name": full_name.strip(),
            "Date of Birth": birth_date.strip(),
            "Age": int(age),            
            "Place of Birth": birth_place.strip()}
```

```python
get_personal_information(text)
```
