### Tokenization in Spacy

Spacy uses object programming in NLP tasks. It provides most efficient NLP algorithm for a given task.

```python
nlp = spacy.load('en_core_web_sm')

doc = nlp('Dr. Strange loves pav ghaji of mumbai. Hulk loves chaat of delhi')
```

```python
# Tokenizing sentences
for sentence in doc.sents:
    print(sentence)
```

```python
# Tokenizing words
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

### Tokenization in NLTK
Is mainy a string procesing library. It provides access to many algorithms and customizations

```python
import nltk
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize('Dr. Strange loves pav ghaji of mumbai. Hulk loves chaat of delhi')
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize('Dr. Strange loves pav ghaji of mumbai. Hulk loves chaat of delhi')
```
