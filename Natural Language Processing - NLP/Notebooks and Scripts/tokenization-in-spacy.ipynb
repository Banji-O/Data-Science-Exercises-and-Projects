### Tokenizer Customization

```python
doc = nlp('gimme double cheese extra large healthy pizza')

tokens = [token.text for token in doc]
tokens
```

```python
# to separate the slang word like 'gimme'

from spacy.symbols import ORTH

nlp.tokenizer.add_special_case('gimme', [
    {ORTH: 'gim'},
    {ORTH: 'me'}
])

doc = nlp('gimme double cheese extra large healthy pizza')

tokens = [token.text for token in doc]
tokens
```

## Exercise

Use spacy to grab all the urls in the provided paragraph.

```python
text = """
Look for data to help you address the question.. Governments are good
sources because data from public research is often freely available. Good
places to start include http://www.data.gov/, and http://www.science.gov/,
and in the United Kingdom, http://data.gov.uk/.
Two of my favorite data sets are General Social Survey at http://www3.norc.org/gss+website/,
and the European Social Survey at http://www.europeansocialsurvey.org/.
"""
```

```python
import spacy

nlp = spacy.blank('en')
```

```python
# Extraction of urls
doc = nlp(text)

urls = []
tokens = [url.text for url in doc if url.like_url]
urls.append(tokens)

urls
```

```python
# Extraction of title words

titles = []
words = [word.text for word in doc if word.text.istitle()]
titles.append(words)

titles
```

```python
# Extraction of title words

titles = []
words = [word.text for word in doc if word.is_title]
titles.append(words)

titles
```

### Extract all money transactions from below sentence along with currency

```python
transactions = 'Tony gave two $ to Peter, Bruce gave 500 £ to Steve'
```

```python
# Process the text
doc = nlp(transactions)

# Extract monetary amounts with currencies
money_with_currency = []

tokens = list(doc)
for i, token in enumerate(tokens):
    # Check if the current token is numeric or a number word
    if token.like_num or token.text.isdigit():
        # Check if the next token is a currency symbol using is_currency
        if i + 1 < len(tokens) and tokens[i + 1].is_currency:
            money_with_currency.append(f"{token.text} {tokens[i + 1].text}")

# Output the results
print(money_with_currency)
```
